# X and Y Creation & Data Splitting

```python
#X and y creation
x = housing_df.drop(columns=['Id'])
y = x.pop("SalePrice")

#Data splitting
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
```

# Pipeline

```python
# select categorical and numerical column names
x_cat_columns = x.select_dtypes(exclude="number").copy().columns
x_num_columns = x.select_dtypes(include="number").copy().columns

# create numerical pipeline, only with the SimpleImputer(strategy="mean")
numeric_pipe = make_pipeline(
    SimpleImputer(strategy="mean"))
 
 # create categorical pipeline, with the SimpleImputer(fill_value="N_A") and the OneHotEncoder
categoric_pipe = make_pipeline(
    SimpleImputer(strategy="constant", fill_value="N_A"),
    OneHotEncoder(sparse_output=False, handle_unknown='ignore')
)

#Import
from sklearn.compose import ColumnTransformer

#Created the columntransformer
preprocessor = ColumnTransformer(
    transformers=[
        ("num_pipe", numeric_pipe, x_num_columns), #1 branch called "num_pipe", will apply the steps in the numeric_pipe to the columns named in x_num_columns
        ("cat_pipe", categoric_pipe, x_cat_columns), #2 branch called "cat_pipe", will apply the steps in the categoric_pipe to the columns named in x_cat_columns
    ]
)

#Creating a pipeline with a scaler 
pipeline = make_pipeline(preprocessor,
                         MinMaxScaler()).set_output(transform='pandas')

pipeline
```

# Iteration 1: No PCA 

```python
start_time = time.time()

pipeline_no_pca = make_pipeline(preprocessor,
                         StandardScaler(),
                         KNeighborsRegressor())

param_grid = {"kneighborsregressor__n_neighbors": list(range(3, 30)),
              "kneighborsregressor__weights":["uniform", "distance"],
              "kneighborsregressor__p":[1,2]}

neigh1_search = RandomizedSearchCV(pipeline_no_pca,
                                  param_grid,
                                  scoring="r2",
                                  n_iter=50,
                                  cv=5,
                                  n_jobs=-1,
                                  random_state=123)

neigh1_search.fit(x_train, y_train)
pred_knn = neigh1_search.predict(X = x_test)

end_time = time.time()

time_taken_no_pca = end_time - start_time
r2_score_no_pca = r2_score(y_test, pred_knn)

print("--- %s seconds ---" % (time_taken_no_pca))
print(f"{(r2_score_no_pca)} R2 Score")
```

# PCA Picking the Number of Components 1: Choosing a Number

```python
#Use the pipeline created above to fit and transform the train data 
x_train_pca_1 = pipeline.fit_transform(x_train, y_train)

#Transform the test set
x_test_pca_1 = pipeline.transform(x_test)
```

```python
from sklearn.decomposition import PCA

#Picking 10 components 
pca = PCA(n_components=10) # pick number of components 
X_reduced = pca.fit_transform(x_train_pca_1)
```

# PCA Picking the Number of Components 2: By Variance

As you saw above, if we are too heavy handed with PCA, and reduce the amount of columns too drastically, we have a massive effect on the accuracy. So, how do we know what the optimal amount of columns are?

We can explore the features using PCA's explained_variance_ratio_ attribute. This highlights the amount of variance explained by each of the selected components. In other words, this will show us how much of the total variance, as a percentage, each of the columns accounts for.

We'll set n_components to None so we can explore the amount of variance of all components.

```python
#Use the pipeline created above to fit and transform the train data 
x_train_pca_2 = pipeline.fit_transform(x_train, y_train)

#Transform the test set
x_test_pca_2 = pipeline.transform(x_test)
```

```python
pca = PCA(n_components=None)
pca.fit(x_train_pca_2)

np.round(pca.explained_variance_ratio_, 4)*100
```

Look at all of the 0's at the end, looks like we have quite a few columns that are providing our model without a lot of information.

We can also visualise cumulative total of pca.explained_variance_ratio_ to see how many of our columns are useful.

```python
import matplotlib.pyplot as plt

cumsum = np.cumsum(pca.explained_variance_ratio_)

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(cumsum, label = "Explained variance");
```

It's common to choose the amount of features that make up 95% of the variance. This way we are feeding our model almost the same amount of data, but reducing the number of features. The cell below calculates the number of features that make up 95% of the variance.

```python
sum(cumsum < 0.95)
```

```python
start_time = time.time()

pipeline_pca_95 = make_pipeline(preprocessor,
                         PCA(n_components=0.95),
                         KNeighborsRegressor())

param_grid = {"kneighborsregressor__n_neighbors": list(range(3, 30)),
              "kneighborsregressor__weights":["uniform", "distance"],
              "kneighborsregressor__p":[1,2]}

neigh3_search = RandomizedSearchCV(pipeline_pca_10,
                                  param_grid,
                                  scoring="r2",
                                  n_iter=50,
                                  cv=5,
                                  n_jobs=-1,
                                  random_state=123)

neigh3_search.fit(x_train, y_train)
pred_knn = neigh3_search.predict(X = x_test)

end_time = time.time()

time_taken_pca_95 = end_time - start_time
r2_score_pca_95 = r2_score(y_test, pred_knn)

print("--- %s seconds ---" % (time_taken_pca_95))
print(f"{(r2_score_pca_95)} R2 Score")
```

# PCA Picking the Number of Components 3: Cross Validation

It is also possible to place different values for PCA in your parameter search to ensure you get the best fit for your data. This may seem counter productive at first, as we are training many pca models, and therefore losing our time gains. However, if you are retraining a model, as you've added some extra data to the dataset, you would just use the previous best value of n_components and retrain the model without full cross validation, and therefore benefit from the time gains.

```python
start_time = time.time()

pipeline_pca_search = make_pipeline(preprocessor,
                                    PCA(),
                                    KNeighborsRegressor())

param_grid = {"kneighborsregressor__n_neighbors": list(range(3, 30)),
              "kneighborsregressor__weights":["uniform", "distance"],
              "kneighborsregressor__p":[1,2]}

neigh4_search = RandomizedSearchCV(pipeline_pca_10,
                                  param_grid,
                                  scoring="r2",
                                  n_iter=50,
                                  cv=5,
                                  n_jobs=-1,
                                  random_state=123)

neigh4_search.fit(x_train, y_train)
pred_knn = neigh4_search.predict(X = x_test)

end_time = time.time()

time_taken_search_pca = end_time - start_time
r2_score_search_pca = r2_score(y_test, pred_knn)

print("--- %s seconds ---" % (time_taken_search_pca))
print(f"{(r2_score_search_pca)} R2 Score")
```

```python
neigh4_search.best_params_
```

```python
start_time = time.time()

pipeline_pca_retrain = make_pipeline(preprocessor,
                                    PCA(n_components=250),
                                    KNeighborsRegressor(weights='uniform',
                                                        p=2,
                                                        n_neighbors=27))

neigh5_search = pipeline_pca_retrain.fit(x_train, y_train)

pred_knn = neigh5_search.predict(X = x_test)

end_time = time.time()

time_taken_retrain_pca = end_time - start_time
r2_score_retrain_pca = r2_score(y_test, pred_knn)

print("--- %s seconds ---" % (time_taken_retrain_pca))
print(f"{(r2_score_retrain_pca)} R2 Score")
```
